# Entiendo la data de Ventas de vehiculos usados

## Descripcion
El conjunto de datos "Ventas de vehículos y tendencias del mercado" ofrece una recopilación exhaustiva de información relativa a las transacciones de venta de diversos vehículos. Este conjunto de datos incluye detalles como el año, la marca, el modelo, la versión, el tipo de carrocería, el tipo de transmisión, el número de identificación del vehículo (VIN), el estado de matriculación, la calificación del estado, el kilometraje, los colores exterior e interior, la información del vendedor, los valores del Informe de Mercado de Manheim (MMR), los precios de venta y las fechas de venta.
